```python
# data splitting 
from sklearn.model_selection import train_test_split
```

```python
# to splitt the main data into train & test set
train_test_split(x,y,test_size=.2, random_state = 7000)
```

```python
#to save the train and test datasets 
x_model, x_extest, y_model, y_extest = train_test_split(x,y,test_size=.2, random_state = 7000)
```

```python
#to check the number of rows at training set 
x_model.shape
```

```python
# to c check the number of rows in testset 
x_extest.shape
```

```python
# to split the modeling set further into test and validation set  
train_test_split(x_model,y_model,test_size=.3, random_state = 7000 )
```

```python
# save the data 
x_train, x_test2, y_train, y_test2 = train_test_split(x_model,y_model,test_size=.3, random_state = 7000)
```

```python
x_train.shape
```

```python
x_test2.shape
```

```python
# further split the test2 into validation and test set 
x_val, x_test, y_val, y_test = train_test_split(x_test2,y_test2,test_size=.5, random_state = 7000)
```

```python
x_val.shape
```

```python
x_test.shape
```

```python
 # import models multilayer perceptrons
from sklearn.neural_network import MLPRegressor
```

```python
#stating the model with default parameter 
nn = MLPRegressor(learning_rate = "constant", activation = "logistic",
          solver ="sgd", hidden_layer_sizes=16, random_state = 70000)
nn.max_iter=10000
```

```python
nn.fit (x_train,y_train)
```

```python
#R2 (coefficient of determination calculation) of # How well the model predictions approximate the true values.
# 1 = perfect fit # Code: print(name of modelstated.score(x_values,y_values_model))
print(nn.score(x_test,y_test))
```

```python
print(nn.score(x_train,y_train))
```

```python
print(nn.score(x_val,y_val))
```

```python
print(nn.score(x_extest,y_extest))
```

```python
#Grid search for different hyperpermeter tuning 
from sklearn.model_selection import GridSearchCV
```

```python
# creatting parameter gride
param_grid_nn= {'learning_rate': ['constant','adaptive','invscaling'], 'activation':['identity', 'logistic', 'tanh', 'relu'], 
    'solver': ['adam', 'sgd', 'lbfgs'], 'hidden_layer_sizes': [5,10,20]}
```

```python
nn_grid = GridSearchCV(estimator =nn, param_grid=param_grid_nn,
                              cv = 5, verbose =2,
                              n_jobs= -1,  scoring = "neg_mean_squared_error")

nn_grid.fit (x,y)
```

```python

nn_grid.best_params_
```
