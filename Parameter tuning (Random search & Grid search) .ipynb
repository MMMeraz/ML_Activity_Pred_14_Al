```python
from sklearn.neural_network import MLPRegressor
```

```python
#now we will state the network with the default setting 
nn = MLPRegressor ()
```

```python
#Instantiate randomized search CV and fit to dataset
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
# creatting parameter gride
param_grid_nn= {'learning_rate': ['constant','adaptive','invscaling'], 'activation':['identity', 'logistic', 'tanh', 'relu'], 
    'solver': ['adm', 'sgd', 'lbfgs'], 'hidden_layer_sizes': [3,10,30]}
```

```python
#Instantiate randomized search CV and fit to dataset
nn_random = RandomizedSearchCV(estimator =nn, param_distributions=param_grid_nn, n_iter = 10000,
                              cv = 5, verbose =2,
                              n_jobs= -1, scoring = "neg_mean_squared_error" )
```

```python
nn_random.fit (x,y)
```

```python
nn_random.best_params_
```

```python
nn_grid = GridSearchCV(estimator =nn, param_grid=param_grid_nn,
                              cv = 5, verbose =2,
                              n_jobs= -1, scoring = "neg_mean_squared_error" )

nn_grid.fit (x,y)
```

```python
nn_grid.best_params_
```
