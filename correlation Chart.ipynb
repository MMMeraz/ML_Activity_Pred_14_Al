```python
# Import seaborn
import seaborn as sns
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(dataset.corr(),annot=True, annot_kws={"size":8}, fmt='.2g', cmap='coolwarm')

  
plt.show()
```

```python
plt.figure(figsize=(12,10))
res=sns.heatmap(dataset.corr(), annot=True, annot_kws={"size":6}, fmt='.2g', cmap='RdBu',vmin=-1, vmax=1,
                cbar=True, cbar_kws=None)

# Drawing the frame
for _, spine in res.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(2)
    spine.set_color('black')
plt.show()
```

```python
plt.figure(figsize=(6,4))
res=sns.heatmap(dataset.corr(), fmt='.2g', cmap=plt.cm.CMRmap_r ,vmin=-1, vmax=1,
                cbar=True, cbar_kws=None)

# Drawing the frame
for _, spine in res.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(1)
    spine.set_color('black')
plt.show()
```

```python
plt.figure(figsize=(10,7))
res=sns.heatmap(dataset.corr(), fmt='.2g',cmap='viridis',  vmin=-1, vmax=1,
                cbar=True, cbar_kws=None)

# Drawing the frame
for _, spine in res.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(1)
    spine.set_color('black')
plt.show()
```
