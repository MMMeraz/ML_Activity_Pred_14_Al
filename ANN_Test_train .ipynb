```python
# Import Section
import csv
import codecs
import sys
import io
import numpy as np
import pandas as pd
import scipy as sp

# For Regression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from statistics import mean


from sklearn.neighbors import KNeighborsRegressor

# Python script for confusion matrix creation. 
from sklearn.metrics import *
from numpy import mean
from numpy import std
from sklearn import metrics

```

```python
dataset=pd.read_excel("Des_cod_1056_92-29.xlsx",index_col=0)
```

```python
dataset.shape
```

```python
# to check the null values 
dataset.isnull().sum()
```

```python
x = dataset.drop(['Act'],axis=1).astype (float)
y = dataset['Act']
```

```python
x.shape
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2, random_state = 47, shuffle=True)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
 # import models multilayer perceptrons
from sklearn.neural_network import MLPRegressor
```

```python
#stating the model with default parameter 
regressor = MLPRegressor(learning_rate = "adaptive", activation = "logistic",
          solver ="lbfgs", hidden_layer_sizes=27, random_state = 42)
regressor.max_iter=10000
```

```python
regressor.fit (x_train,y_train)
```

```python
# compute the testing accuracy
y_pred = regressor.predict(x_test)
```

```python
y_pred_train = regressor.predict(x_train)
```

```python
# parameter tuning grid searchCV
from sklearn.model_selection import GridSearchCV
```

```python
learning_rate = ['constant','adaptive','invscaling']
activation = ['identity', 'logistic', 'tanh', 'relu']
solver= ['adam', 'sgd', 'lbfgs']
hidden_layer_sizes = [25, 27, 30, 35, 40, 45, 50, 55, 60, 70, 80, 90, 100, 110, 120, 130, 150]
```

```python
param_grid_regressor1= {'learning_rate':learning_rate, 'activation':activation, 
    'solver': solver, 'hidden_layer_sizes': hidden_layer_sizes}
```

```python
print (param_grid_regressor1)
```

```python
regressor_grid1 = GridSearchCV( estimator = regressor, param_grid = param_grid_regressor1,
                              cv = 5, verbose = 2, n_jobs= -1, scoring = "neg_mean_squared_error")
```

```python
regressor_grid1.fit(x_train, y_train)
```

```python
regressor_grid1.best_params_
```

```python
regressor_grid1.best_score_
```

```python
#Hyperperameters tuning 
hyperparameters = dict(learning_rate=learning_rate, activation=activation, solver= solver, 
                       hidden_layer_sizes= hidden_layer_sizes )
```

```python
clf = GridSearchCV(regressor, hyperparameters, cv=5)
```

```python
best_model = clf.fit(x_train,y_train)
```

```python
print('Best learning_rate:', best_model.best_estimator_.get_params()['learning_rate'])
print('Best activation:', best_model.best_estimator_.get_params()['activation'])
print('Best solver:', best_model.best_estimator_.get_params()['solver'])
print('hidden_layer_sizes:', best_model.best_estimator_.get_params()['hidden_layer_sizes'])
```
